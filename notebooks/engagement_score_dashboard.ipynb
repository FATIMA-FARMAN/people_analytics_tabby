# Engagement Score Dashboard Prep
Summaries for Looker/Power BI and quick plots.

```python
import pandas as pd
import matplotlib.pyplot as plt

eng = pd.read_csv('../data/engagement.csv')
summary = eng.groupby('Department')['Engagement_Score'].agg(['count','mean','median']).sort_values('mean', ascending=False)
summary
```

## Boxplot of Engagement by Department

```python
eng = pd.read_csv('../data/engagement.csv')
groups = [g['Engagement_Score'].values for _, g in eng.groupby('Department')]
labels = [d for d,_ in eng.groupby('Department')]
plt.boxplot(groups, labels=labels, showmeans=True)
plt.title('Engagement Score by Department')
plt.ylabel('Score')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```
