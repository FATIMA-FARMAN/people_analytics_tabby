# Hiring Trends Analysis
This notebook explores time-to-fill, diversity, and basic hiring metrics using the sample `employees.csv`.

```python
import pandas as pd
import matplotlib.pyplot as plt

employees = pd.read_csv('../data/employees.csv')
employees.head()
```

## Time-to-Fill by Department

```python
dept_ttf = employees.groupby('Department')['Time_to_Fill_Days'].mean().sort_values()
ax = dept_ttf.plot(kind='bar')
ax.set_title('Average Time-to-Fill (Days) by Department')
ax.set_ylabel('Days')
plt.tight_layout()
plt.show()
```

## Diversity Ratio (Gender Share)

```python
gender_share = employees['Gender'].value_counts(normalize=True)
ax = gender_share.plot(kind='pie', autopct='%1.1f%%')
ax.set_title('Gender Distribution (All Employees)')
ax.set_ylabel('')
plt.tight_layout()
plt.show()
```
